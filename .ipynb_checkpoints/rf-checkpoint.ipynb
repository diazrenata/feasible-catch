```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
orig_samples = pd.read_csv('https://raw.githubusercontent.com/diazrenata/feasible-catch/main/keras/fs_data.csv')
orig_samples.tail()
```

```python
dat = orig_samples.copy()
dat = dat.drop(columns = 'sd')
dat = dat.rename(columns = {"mean": "mean_hill1"})
dat.tail()

```

```python
# Not sure what this reshaping process is accomplishing...

x = dat.richness.values.reshape(-1, 1)
y = dat.mean_hill1.values.reshape(-1, 1)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
train_dat = dat.sample(frac = .8, replace = False, random_state = 0)
```

```python
test_dat = dat.drop(train_dat.index)
```

```python
train_x = train_dat.drop(columns = 'mean_hill1')
train_y = train_dat['mean_hill1']

test_x = test_dat.drop(columns = 'mean_hill1')
test_y = test_dat['mean_hill1']
```

```python
rfrm = RandomForestRegressor()
rfrm.fit(train_x, train_y)
```

```python
y_pred = rfrm.predict(test_x)
y_pred
```

```python
plt.scatter(test_y, y_pred, color = 'green')
plt.plot(test_y, test_y)
```

```python
rfrm.score(test_x, test_y)
```

## Repeating with the newer samples...



```python
new_dat_raw = pd.read_csv("https://raw.githubusercontent.com/diazrenata/feasible-catch/main/new_samples.csv")
```

```python
newdat = new_dat_raw.copy()
newdat = newdat.drop(columns = 'sd')
newdat = newdat.rename(columns = {"mean": "mean_hill1"})
newdat.tail()

```

```python
new_train_dat = newdat.sample(frac = .8, replace = False, random_state = 0)
new_test_dat = newdat.drop(new_train_dat.index)


new_train_x = new_train_dat.drop(columns = 'mean_hill1')
new_train_y = new_train_dat['mean_hill1']

new_test_x = new_test_dat.drop(columns = 'mean_hill1')
new_test_y = new_test_dat['mean_hill1']
```

```python
new_rfrm = RandomForestRegressor()
new_rfrm.fit(new_train_x, new_train_y)
```

```python
new_y_pred = new_rfrm.predict(new_test_x)
new_y_pred
```

```python
plt.scatter(new_test_y, new_y_pred, color = 'green')
plt.plot(new_test_y, new_test_y)
```

```python
new_rfrm.score(new_test_x, new_test_y)
```
